### 预处理

```python
df1=df.iloc[:,[1,2,3]]
q1=df1.quantile(0.25)
q3=df1.quantile(0.75)
iqr=q3-q1
```

```python
index0=df1[~((df1 < (q1 - 1.5 * iqr)) |(df1 > (q3 + 1.5 * iqr))).any(axis=1)].index
```

```python
df=df.iloc[index0,:]
```

```python
df=df.drop(columns=['time_duration','trip_distance','speed'])
```

### cell id匹配

```python
#坐标与cell编号匹配函数
def linear_referencing(locpair):
    x_min = 103.605414
    y_max = 1.470763
    lon_diff = 0.00269582
    lat_diff = 0.00271293
    cell_id = math.floor((locpair[1]-x_min)/lon_diff) + 180 * math.floor((y_max-locpair[0])/lat_diff)
    return cell_id
```

```python
df['cell_start']=df.apply(lambda x:linear_referencing((x['lat0'],x['lng0'])),axis=1)
df['cell_end']=df.apply(lambda x:linear_referencing((x['lat1'],x['lng1'])),axis=1)
df=df.drop(columns=['lng0','lat0','lng1','lat1'])
```

```python
from datetime import date, timedelta

sdate = date(2017, 9, 30)   # start date
edate = date(2017, 11, 1)   # end date

delta = edate - sdate       # as timedelta

days=[]
for i in range(delta.days + 1):
    day = sdate + timedelta(days=i)
    days.append(day)
```

```python
##温度
import urllib
import json

#stations=['S109','S117','S50','S107','S43','S108','S44','S121','S106','S111','S122','s60','S115','S24','S116','S104','S100']
res=[]

for day in days:
    #获取数据
    url = 'https://api.data.gov.sg/v1/environment/air-temperature?date={}'.format(str(day))
    #headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'}
    req=urllib.request.Request(url=url)
    fileobj = urllib.request.urlopen(req).read()
    
    #处理文件格式
    data=json.loads(fileobj.decode("utf-8"))
    
    #寻找station
    stations=[each['id'] for each in data['metadata']['stations']]
    dic={}
    dic['timestamp']=[each['timestamp'] for each in data['items'] if len(each['readings'])==len(stations)]
    for station in stations:
        dic[station]=[value['value'] for each in data['items'] for value in each['readings'] if value['station_id']==station and len(each['readings'])==len(stations)]
        
    res.append(pd.DataFrame(dic))
```

```python
for i in range(len(days)):
    res[i].to_csv('tempo/{}.csv'.format(str(days[i])),index=False)
```

```python
##降雨
res1=[]

for day in days:
    #获取数据
    url = 'https://api.data.gov.sg/v1/environment/rainfall?date={}'.format(str(day))
    #headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'}
    req=urllib.request.Request(url=url)
    fileobj = urllib.request.urlopen(req).read()
    
    #处理文件格式
    data=json.loads(fileobj.decode("utf-8"))
    
    #寻找station
    stations=[each['id'] for each in data['metadata']['stations']]
    dic={}
    dic['timestamp']=[each['timestamp'] for each in data['items'] if len(each['readings'])==len(stations)]
    for station in stations:
        dic[station]=[value['value'] for each in data['items'] for value in each['readings'] if value['station_id']==station and len(each['readings'])==len(stations)]
        
    res1.append(pd.DataFrame(dic))
```

```python
for i in range(len(days)):
    res1[i].to_csv('rainfall/{}.csv'.format(str(days[i])),index=False)
```

## 站点信息

```python
stationinfo={'id':[],'latitude':[],'longitude':[]}

for day in days:
    #获取数据
    url = 'https://api.data.gov.sg/v1/environment/air-temperature?date={}'.format(str(day))
    #headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'}
    req=urllib.request.Request(url=url)
    fileobj = urllib.request.urlopen(req).read()
    
    #处理文件格式
    data=json.loads(fileobj.decode("utf-8"))
    
    #记录station
    for station in data['metadata']['stations']:
        if station['id'] not in stationinfo['id']:
            stationinfo['id'].append(station['id'])
            stationinfo['latitude'].append(station['location']['latitude'])
            stationinfo['longitude'].append(station['location']['longitude'])
```

```python
comb1=[] # 19
comb2=[] # 18

for day in days:
    temp=pd.read_csv('tempo/{}.csv'.format(str(day)))
    flow1=flow[flow['date']==str(day)]
    
    temp['hour']=pd.to_datetime(temp['timestamp']).dt.hour
    if len(temp):
        temp=temp.groupby(by='hour').median().reset_index()
    else:
        continue

    alldata=flow1.merge(temp,left_on='start_hour',right_on='hour',how='inner')
      
    if len(temp.columns)==19:
        comb1.append(alldata)
    else:
        comb2.append(alldata)
        
tempo19=pd.concat(comb1,axis=0)
tempo18=pd.concat(comb2,axis=0)
```

```python
##合并
import os
def find_all(name, path):
    result = []
    for root, dirs, files in os.walk(path):
        if name in files:
            result.append(os.path.join(root, name))
    return result
```

## 月温度变化

```python
tempre=[]
for day in days:
    temp = pd.read_csv('tempo/{}.csv'.format(str(day)))
    if len(temp):
        temp['hour'] = pd.to_datetime(temp['timestamp']).dt.hour
        temp['median'] = temp.iloc[:,1:].median(axis=1)
        temp = temp[['hour','median']].groupby('hour').median().reset_index()
        temp['date'] = day
        tempre.append(temp)
```

```python
tempre = pd.concat(tempre)
```

```python
tempre = tempre.reset_index(drop=True)
tempre['date'] = tempre['date'].astype(str)
```
